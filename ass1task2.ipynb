```python
import matplotlib.pyplot as plt
import pandas as pd
people = ['numbu', 'pani', 'puri', 'attu']
age =[30,40,21,20]
height =[100,200,178,171]
weight=[50,40,70,100]

```

```python
data=pd.DataFrame({
'name':people,
'Age':age ,
'Height':height,
'Weight':weight
})
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Age'], data['Weight'], color='red')
plt.title('Scatter Plot of Age vs Weight')
plt.xlabel('Age')
plt.ylabel('Weight')
```

```python
plt.figure(figsize=(8, 6))
plt.bar(data['name'], data['Height'], color='green')
plt.title('Bar Chart of Height for Each Person')
plt.xlabel('Name')
plt.ylabel('Height')
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['Weight'], bins=10, color='orange', edgecolor='black')
plt.title('Histogram of Weight Distribution')
plt.xlabel('Weight')
plt.ylabel('Frequency')
```
